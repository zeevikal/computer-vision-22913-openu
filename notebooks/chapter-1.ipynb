### #2 Page 13 - smoothing

```python
im = misc.imread('../data/cameraman.tiff')
```

```python
plt.imshow(im)
```

```python
def my_smooth(img):
    sm_img = []
    for i in range(1, img.shape[0] - 1):
        sm_img_row = []
        for j in range(1, img.shape[1] - 1):
            sm_img_row.append(((1/9) * (img[i-1][j-1]) + (img[i-1][j]) + (img[i-1][j+1]) + (img[i][j-1]) + 
                           (img[i][j]) + (img[i][j+1]) + (img[i+1][j-1]) + (img[i+1][j]) + (img[i+1][j+1])))
        sm_img.append(sm_img_row)
    return np.array(sm_img)
```

```python
im_smt = my_smooth(im)
plt.imshow(im_smt)
```


### #3 Page 13 - negative image

```python
im = misc.imread('../data/cameraman.tiff')
```

```python
def negative_img(img):
    neg_img = []
    for i in range(1, img.shape[0] - 1):
        neg_img_row = []
        for j in range(1, img.shape[1] - 1):
            neg_img_row.append(255 - img[i][j])
        neg_img.append(neg_img_row)
    return np.array(neg_img)
```


### #4 Page 14 - save image as binary file

```python
def im_to_file(im_path):
    im_array = []
    im = misc.imread(im_path)
    r = im.shape[0]
    c = im.shape[1]
    im_array.append(r)
    im_array.append(c)
    for i in range(0, r):
        for j in range(0, c):
            im_array.append(im[i][j])
    np.save('../data/im_bin', np.array(im_array))
```

```python
im_to_file('../data/cameraman.tiff')
```

### #5 Page 14 - image format replacement 

```python
def change_img_format(img_path):
    im = misc.imread(img_path)
    if '.tiff' in img_path:
        misc.imsave('../data/out_img.bmp', im)
    elif '.bmp' in img_path:
        misc.imsave('../data/out_img.tiff', im)
```

### #6 Page 15 - threshold

```python
def threshold_img(img_path, t):
    im = misc.imread(img_path)
    im_t = im < t
    im_t = im_t.astype(int)
    misc.imsave('../data/threshold_img.tiff', im_t)
    return im_t
```

### #7 Page 17 - half toning

```python
def ht_color_pixel(ht_img, i, j, pix_arr):
    ht_img[2*i][2*j] = pix_arr[0]
    ht_img[2*i + 1][2*j] = pix_arr[1]
    ht_img[2*i][2*j + 1] = pix_arr[2]
    ht_img[2*i + 1][2*j + 1] = pix_arr[3]
    return ht_img

def half_tone_img(img_path):
    im = misc.imread(img_path)
    r = im.shape[0]
    c = im.shape[1]
    ht_img = np.empty([2*r, 2*c])
    for i in range(0, r):
        ht_row = []
        for j in range(0, c):
            if 0 <= im[i][j] <= 50:
                ht_img = ht_color_pixel(ht_img, i, j, [0, 0, 0, 0])
            elif 51 <= im[i][j] <= 101:
                ht_img = ht_color_pixel(ht_img, i, j, [0, 0, 255, 0])
            elif 102 <= im[i][j] <= 152:
                ht_img = ht_color_pixel(ht_img, i, j, [0, 255, 0, 255])
            elif 153 <= im[i][j] <= 203:
                ht_img = ht_color_pixel(ht_img, i, j, [255, 0, 0, 0])
            elif 204 <= im[i][j] <= 255:
                ht_img = ht_color_pixel(ht_img, i, j, [0, 0, 0, 0])
    return ht_img            
```

```python
ht_img = half_tone_img('../data/cameraman.tiff')
plt.imshow(ht_img)
```

### #8 Page 18 - error diffusion

```python
def error_diffusion(img_path, t):
    im = misc.imread(img_path)
    for i in range(0, im.shape[0] - 1):
        for j in range(0, im.shape[1] - 1):
            old_pixel = im[i][j]
            im[i][j] = 255 if im[i][j] > t else 0
            error = old_pixel - im[i][j]
            im[i][j + 1] = im[i][j + 1] + (3 * error) // 8
            im[i + 1][j + 1] = im[i + 1][j + 1] + error // 4
            im[i + 1][j] = im[i + 1][j] + (3 * error) // 8
    return im
```

```python
im = error_diffusion('../data/cameraman.tiff', 155)
plt.imshow(im)
```

### #11 Page 23 - distance function

```python
# f((x, y)(u, v)) = (x - u)^2 + (y - v)^2 - is it a distance function according those 4 rules?
# 1. f(x, y) >= 0
# 2. f(x, x) = 0
# 3. f(x, y) = f(y, x)
# 4. if x, y, z are dots in a same area so f(x, y) + f(y, z) >= f(x, z)

x = 1
y = 4
u = -3
v = 8

def if_dis_func(x, y, u, v):
    zx = 2
    zy = 8
    if ((x-u)^2 + (y-v)^2) < 0:
        print('Not because rule no. 1')
        return False
    elif ((x-x)^2 + (y-y)^2) != 0:
        print('Not because rule no. 2')
        return False
    elif ((x-u)^2 + (y-v)^2)  == ((u-x)^2 + (v-y)^2):
        print('Not because rule no. 3')
        return False
    elif (((x-u)^2 + (y-v)^2) + ((u-zx)^2 + (v-zy)^2)) < 0:
        print('Not because rule no. 4')
        return False
    return True

if_dis_func(x, y, u, v)
```

### #12 Page 24 - unit circle

```python
# x^2 + y^2 = 1

def eucledian_circle(x,y):
    return np.sqrt(x*x+y*y)

def block_circle(x,y):
    return (np.abs(x) + np.abs(y))

def chess_circle(x,y):
    return np.maximum(np.abs(x),np.abs(y))

xx=np.linspace(-2,2,400)
yy=np.linspace(-2,2,400)
[X,Y]=np.meshgrid(xx,yy)
```

```python
Z=eucledian_circle(X,Y)

plt.figure()
plt.contour(X,Y,Z,[1])
plt.show()
```

```python
Z=block_circle(X,Y)

plt.figure()
plt.contour(X,Y,Z,[1])
plt.show()
```

```python
Z=chess_circle(X,Y)

plt.figure()
plt.contour(X,Y,Z,[1])
plt.show()
```
