```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
%matplotlib inline
```

### #1 Page 10

```python
# 1.
lin1 = misc.imread('../data/cameraman.tiff')
plt.imshow(lin1)
```

```python
lout1 = []
for i in range(0, lin1.shape[0]):
    if i % 2 is 0:
        lout1_row = []
        for j in range(0, lin1.shape[1]):
            if j % 2 is 0:
                lout1_row.append(lin1[i][j])
        lout1.append(lout1_row)
```

```python
plt.imshow(lout1)
```

```python
# 2.
lout2 = []
for i in range(0, len(lout1[0])):
    lout2_row = []
    for j in range(0, len(lout1[1])):
        for k in range(0, 2):
            lout2_row.append(lout1[i][j])
    for k in range(0, 2):
        lout2.append(lout2_row)
```

```python
plt.imshow(lout2)
```

```python
# 3.
f, axarr = plt.subplots(1, 2)
axarr[0].imshow(lin1)
axarr[1].imshow(lout2)
```
