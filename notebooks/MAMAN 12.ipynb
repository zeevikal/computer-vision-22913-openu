```python
im = misc.imread('../data/cameraman.tiff')
plt.imshow(im)
```

```python
w, h = im.shape
```

```python
kernel_1 = np.array([[-1, -1, -1], [-1, 10, -1], [-1, -1, -1]])
```

```python
def filter_img(img_path, kernel):
    img = misc.imread(img_path)
    w, h = img.shape
    filtered_img = np.zeros(img.shape)
    for i in range(1,w-3):
            for j in range(2, h-3):
                neigh_img = img[i-1:i+2, j-1:j+2]
                kernel_output = np.sum(neigh_img*kernel)
                filtered_img[i, j] = kernel_output
    return filtered_img
```

```python
plt.imshow(filter_img('../data/cameraman.tiff', kernel_1))
```

```python
kernel_2 = np.array([[-1, -1, -1], [-1, -10, -1], [-1, -1, -1]])
plt.imshow(filter_img('../data/cameraman.tiff', kernel_2))
```

as we can see, if our kernel will be negative the image we'll get lower values, till it will be a negative image of the original one.
