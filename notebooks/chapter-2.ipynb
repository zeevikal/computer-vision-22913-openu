```python
def sinsum(x, k):
    out = np.zeros(len(x))
    for n in range(1, k, 2):
        out += 4 * np.sin(2 * np.pi * n * x) / np.pi * n
    return out
```

```python
plt.plot(x, sinsum(x, 2))
```

```python
plt.plot(x, sinsum(x, 10))
```

```python
plt.plot(x, sinsum(x, 100))
```
